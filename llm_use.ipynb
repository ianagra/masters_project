```python
import pandas as pd
import json
```

```python
# Função para converter o DataFrame em JSON ou formato legível para o Langchain
def preprocess_data(df):
    records = df.to_dict(orient="records")
    json_data = json.dumps(records, indent=4)
    return json_data
```

```python
df_surv = pd.read_csv("dataset_survival_thr_d.csv")
df_coefficients = pd.read_csv("coefficients_thr_d.csv")
```

```python
# Convertendo os dados para JSON
survival_json = preprocess_data(df_surv)
print(survival_json)
coefficients_json = preprocess_data(df_coefficients)
print(coefficients_json)
```
